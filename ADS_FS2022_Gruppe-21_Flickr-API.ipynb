## <font color="blue"> Flickr REST API </font>
### <font color="blue"> Download "Fahrrad" images </font>
<font color="blue">Call the flickr API with the provided keys, search for keyword "Fahrrad" and download all Images that are medium large (url_c). 
Define the number of derived urls with changing the "#Get number urls" amount</font>

```python
FLICKR_PUBLIC = 'c8cf6de83798c6aa3bf7ff1bc3efc67e'
FLICKR_SECRET = '36f711532617cf85'

flickr=flickrapi.FlickrAPI('c8cf6de83798c6aa3bf7ff1bc3efc67e', '36f711532617cf85', cache=True)

keyword = 'fahrrad'

photos = flickr.walk(text=keyword,
                     tag_mode='all',
                     tags=keyword,
                     extras='url_c',
                     per_page=1000,           
                     sort='relevance')

# Initialize list
urls = []

# Derive urls
for i, photo in enumerate(photos):
    url = photo.get('url_c')
    urls.append(url)
    # print(urls[i])
    # Get number urls
    if i > 1000:
        break
        
# Filter 'None' values from list
urls_clean = list(filter(None, urls)) 
urls_clean

# Print cleaned list
for i, url in enumerate(urls_clean):
        print(urls_clean[i])
```

<font color="blue">Download the defined number of images into a folder that will be automatically created and named after your keyword. Then show (plot) a selection of the downloaded photos</font>

```python
os.makedirs(keyword, exist_ok=True)
os.chdir(os.path.join(path, "gruppe21-data", keyword))
os.getcwd()
```

```python
for i, url in enumerate(urls_clean):    
    urllib.request.urlretrieve(url,  str('photo_' + str(i+1) + '.jpg'))
```

```python
#Plot 5 images to check the quality and content
%matplotlib inline

for i, url in enumerate(urls_clean[0:5]):
    img = mpimg.imread(str('photo_' + str(i+1) + '.jpg'))
    plt.figure(figsize=(8,8))
    plt.axis('off')
    imgplot = plt.imshow(img)
```
